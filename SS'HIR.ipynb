# Direct fit with SHIR

### Direct

```python
def func(t, b_eff, mu, sigma):
    return -(sigma + mu) + b_eff * np.exp(-sigma*t)
```

```python
fits = []

for i,m in enumerate(mod_new):
    bounds=([0.,0.,0.],[np.inf,np.inf,np.inf])
    params, params_covariance = optimize.curve_fit(func, np.arange(m.T_va), m.a_data, bounds=bounds)
    fits.append(params)
    print(params)
```

```python
plt.figure(figsize=(14,12))

for i,m in enumerate(mod_new):
    plt.subplot(2,2,i+1)
    
    t = np.arange(m.T_va)
    plt.plot(t, func(t, fits[i][0], fits[i][1], fits[i][2]), label='Fit model')
    
    plt.plot(t, m.a_data, label='Exp. data')
    
    #plt.xlim([-7., -3.5])
    plt.axhline(y=0, color='gray', linestyle='-.')
    plt.legend()
    plt.xlabel('days', size=15)
    plt.ylabel('a', size=15)
    plt.title(m.country, size=18)
    
plt.show()
```

```python
opts1 = pd.DataFrame(fits, columns=['beta_eff', 'mu', 'sigma'], index=new_countries)
opts1
```

# Direct fit with SS'HIR

** We can assume that the initial fraction of infected is small, so $S+S'\approx1$.
In this way, $S'=p$, $S'=1-p$

```python
def direct_SSHIR(t, beta, sigma, mu, p):
    return beta*( 1.-p + p*np.exp(-sigma*t) ) - sigma/( (1.-p)/p*np.exp(sigma*t) + 1.) - mu
```

### Direct

```python
fits = []

for i,m in enumerate(mod_new):
    try:
        bounds=([0.,0.,0.,0.],[np.inf,np.inf,np.inf,1.0])
        params, params_covariance = optimize.curve_fit(direct_SSHIR, np.arange(m.T_va), m.a_data, bounds=bounds)
    except:
        bounds=([0.,0.,0.,0.],[np.inf,np.inf,np.inf,np.inf])
        params, params_covariance = optimize.curve_fit(direct_SSHIR, np.arange(m.T_va), m.a_data, bounds=bounds)
    fits.append(params)
    print(params)
```

```python
plt.figure(figsize=(14,12))

for i,m in enumerate(mod_new):
    plt.subplot(2,2,i+1)
    
    t = np.arange(m.T_va)
    plt.plot(t, direct_SSHIR(t, fits[i][0], fits[i][1], fits[i][2], fits[i][3]), label='Fit model')
    
    plt.plot(t, m.a_data, label='Exp. data')
    
    #plt.xlim([-7., -3.5])
    plt.axhline(y=0, color='gray', linestyle='-.')
    plt.legend()
    plt.xlabel('days', size=15)
    plt.ylabel('a', size=15)
    plt.title(m.country, size=18)
    
plt.show()
```

```python
opts1 = pd.DataFrame(fits, columns=['beta', 'sigma', 'mu', 'p'], index=new_countries)
opts1
```

### Mu fixed

```python
def direct_SSHIR_mu_fixed(t, beta, sigma, p):
    mu = 1./14.
    return beta*( 1.-p + p*np.exp(-sigma*t) ) - sigma*(p*np.exp(-sigma*t))/(1.-p + p*np.exp(-sigma*t)) - mu
```

```python
fits = []

for i,m in enumerate(mod_new):
    try:
        bounds=([0.,0.,0.],[np.inf,np.inf,1.0])
        params, params_covariance = optimize.curve_fit(direct_SSHIR_mu_fixed, np.arange(m.T_va), m.a_data, bounds=bounds)

    except:
        bounds=([0.,0.,0.],[np.inf,np.inf,np.inf])
        params, params_covariance = optimize.curve_fit(direct_SSHIR_mu_fixed, np.arange(m.T_va), m.a_data, bounds=bounds)

    fits.append(params)
    print(params)
```
